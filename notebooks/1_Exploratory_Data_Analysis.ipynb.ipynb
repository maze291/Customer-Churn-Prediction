```python
# Customer Churn Prediction - Exploratory Data Analysis

## Setup and Data Loading

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Visualization style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# File paths
relative_paths = ['../data/raw/churn-bigml-80.csv', '../data/raw/churn-bigml-20.csv']
absolute_paths = [
    r"C:\Users\maze2\Customer-Churn-Prediction\data\raw\churn-bigml-80.csv",
    r"C:\Users\maze2\Customer-Churn-Prediction\data\raw\churn-bigml-20.csv"
]

# Try to load the datasets
try:
    # First try relative paths
    train_data = pd.read_csv(relative_paths[0])
    test_data = pd.read_csv(relative_paths[1])
    print("Succesfully loaded in data")
except FileNotFoundError:
    try:
        # If relative paths fail, try absolute paths
        train_data = pd.read_csv(absolute_paths[0])
        test_data = pd.read_csv(absolute_paths[1])
        print("Successfully loaded data using absolute paths.")
    except FileNotFoundError:
        print("ERROR: Could not find the CSV files!")
        print("Check files")
        print(f"  - Relative: {relative_paths[0]}")
        print(f"  - Absolute: {absolute_paths[0]}")

print(f"Training set shape: {train_data.shape}")
print(f"Test set shape: {test_data.shape}")

## Data Overview

# Display first few rows of the training dataset
print("\nFirst few rows of the training dataset:")
display(train_data.head())

# Examine data types and missing values
print("\nData types and missing values:")
train_data.info()

# Summary statistics
print("\nSummary statistics:")
display(train_data.describe())

## Target Variable Distribution

# Check the distribution of the target variable (Churn)
churn_distribution = train_data['Churn'].value_counts(normalize=True) * 100
print("\nChurn Distribution in Training Data:")
print(churn_distribution)

# Verify churn is Boolean type
print(f"\nChurn data type: {train_data['Churn'].dtype}")

# Visualize the churn distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Churn', data=train_data)
plt.title('Distribution of Churn')
plt.ylabel('Count')
plt.show()

# Compare churn distribution in training and test sets
plt.figure(figsize=(10, 6))
train_churn = train_data['Churn'].value_counts(normalize=True)
test_churn = test_data['Churn'].value_counts(normalize=True)

# converting boolean index to text for more readable labels
train_churn.index = ['Not Churned', 'Churned'] if not train_churn.index[0] else ['Churned', 'Not Churned']
test_churn.index = ['Not Churned', 'Churned'] if not test_churn.index[0] else ['Churned', 'Not Churned']

labels = ['Not Churned', 'Churned']
x = np.arange(len(labels))
width = 0.35

plt.bar(x - width/2, train_churn, width, label='Train')
plt.bar(x + width/2, test_churn, width, label='Test')
plt.title('Churn Distribution: Train vs Test')
plt.xticks(x, labels)
plt.legend()
plt.show()

## Numerical Features Analysis

# Identify numerical features
numerical_features = train_data.select_dtypes(include=['int64', 'float64']).columns.tolist() #convert 
numerical_features = [col for col in numerical_features if col != 'Churn']

print(f"\nNumber of numerical features: {len(numerical_features)}")
print("Numerical features:", numerical_features)

# Visualize distributions of numerical features
for feature in numerical_features[:5]:  # First 5 numerical features
    plt.figure(figsize=(12, 5))
    
    # Distribution plot
    plt.subplot(1, 2, 1)
    sns.histplot(train_data[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    
    # Box plot by churn status
    plt.subplot(1, 2, 2)
    sns.boxplot(x='Churn', y=feature, data=train_data)
    plt.title(f'{feature} by Churn Status')
    
    plt.tight_layout()
    plt.show()

# Correlation between numerical features
correlation_matrix = train_data[numerical_features].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

## Categorical Features Analysis

# Identify categorical features
categorical_features = train_data.select_dtypes(include=['object']).columns.tolist()
print(f"\nNumber of categorical features: {len(categorical_features)}")
print("Categorical features:", categorical_features)

# Examine categorical features
for feature in categorical_features:
    plt.figure(figsize=(12, 5))
    
    # Value counts ie Unique value frequencies
    counts = train_data[feature].value_counts()
    print(f"\n{feature} distribution:")
    print(counts)
    
    # Visualize distribution
    plt.subplot(1, 2, 1)
    sns.countplot(y=feature, data=train_data, order=counts.index)
    plt.title(f'Distribution of {feature}')
    
    # Distribution by churn(True not Yes)
    plt.subplot(1, 2, 2)
    churn_rate = pd.crosstab(train_data[feature], train_data['Churn'], normalize='index')
    # Use True instead of 'Yes' since Churn is boolean
    churn_rate[True].sort_values().plot(kind='barh') #horizontal bar chart
    plt.title(f'Churn Rate by {feature}')
    plt.xlabel('Churn Rate')
    
    plt.tight_layout()
    plt.show()

## Key Insights and Next Steps

print("\n## Key Insights from EDA:")
print("1. Class Distribution: Significant imbalance with ~85% non-churned vs ~15% churned customers")
print("2. Important Features: Customer service calls and International plan appear to be strong predictors")
print("3. Correlations: Strong correlation between minutes and charges across all time periods")
print("4. Categorical Variables: International plan shows significant impact on churn rates")
```
